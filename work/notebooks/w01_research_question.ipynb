## 2. The question: decision, action, cost of a wrong call

### **Research Question**
Which pages should be prioritized for content refresh to improve future search performance?


### **Unit of Analysis**
Individual web pages.

### **Output**
A ranked list of pages with an opportunity score.

### **Decision**
Decide which pages should be refreshed first.

### **Action**
SEO teams update the highest-ranked pages before spending time on lower-priority pages.

### **Cost of a Wrong Recommendation**
If high-potential pages are ignored, valuable traffic opportunities may be missed. If low-impact pages are prioritized, time and resources may be wasted.

### **Why ML?**
Multiple factors such as impressions, CTR, page age, search position, and update history interact in ways that are difficult to evaluate consistently using manual rules. Machine learning can help identify useful patterns and prioritize pages more effectively.
